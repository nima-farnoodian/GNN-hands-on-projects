```python
import torch

def load_node_csv(path, index_col, encoders=None, **kwargs):
    df = pd.read_csv(path, index_col=index_col, **kwargs)
    mapping = {index: i for i, index in enumerate(df.index.unique())}

    x = None
    if encoders is not None:
        xs = [encoder(df[col]) for col, encoder in encoders.items()]
        x = torch.cat(xs, dim=-1)

    return x, mapping
```

```python
class SequenceEncoder(object):
    def __init__(self, model_name='all-MiniLM-L6-v2', device=None):
        self.device = device
        self.model = SentenceTransformer(model_name, device=device)

    @torch.no_grad()
    def __call__(self, df):
        x = self.model.encode(df.values, show_progress_bar=True,
                              convert_to_tensor=True, device=self.device)
        return x.cpu()
```

```python
class GenresEncoder(object):
    def __init__(self, sep='|'):
        self.sep = sep

    def __call__(self, df):
        genres = set(g for col in df.values for g in col.split(self.sep))
        mapping = {genre: i for i, genre in enumerate(genres)}

        x = torch.zeros(len(df), len(mapping))
        for i, col in enumerate(df.values):
            for genre in col.split(self.sep):
                x[i, mapping[genre]] = 1
        return x
```

## Extracting links

```python
def load_edge_csv(path, src_index_col, src_mapping, dst_index_col, dst_mapping,
                  encoders=None, **kwargs):
    df = pd.read_csv(path, **kwargs)

    src = [src_mapping[index] for index in df[src_index_col]]
    dst = [dst_mapping[index] for index in df[dst_index_col]]
    edge_index = torch.tensor([src, dst])

    edge_attr = None
    if encoders is not None:
        edge_attrs = [encoder(df[col]) for col, encoder in encoders.items()]
        edge_attr = torch.cat(edge_attrs, dim=-1)

    return edge_index, edge_attr
```

```python
class IdentityEncoder(object):
    def __init__(self, dtype=None):
        self.dtype = dtype

    def __call__(self, df):
        return torch.from_numpy(df.values).view(-1, 1).to(self.dtype)
```

## Link Prediction using SAGEConv

```python
class GNNEncoder(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv((-1, -1), hidden_channels)
        self.conv2 = SAGEConv((-1, -1), out_channels)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index)
        return x
```

```python
class EdgeDecoder(torch.nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        self.lin1 = Linear(2 * hidden_channels, hidden_channels)
        self.lin2 = Linear(hidden_channels, 1)

    def forward(self, z_dict, edge_label_index):
        row, col = edge_label_index
        z = torch.cat([z_dict['user'][row], z_dict['movie'][col]], dim=-1)

        z = self.lin1(z).relu()
        z = self.lin2(z)
        return z.view(-1)
```

```python
class Model(torch.nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        self.encoder = GNNEncoder(hidden_channels, hidden_channels)
        self.encoder = to_hetero(self.encoder, data.metadata(), aggr='sum')
        self.decoder = EdgeDecoder(hidden_channels)

    def forward(self, x_dict, edge_index_dict, edge_label_index):
        z_dict = self.encoder(x_dict, edge_index_dict)
        return self.decoder(z_dict, edge_label_index)
```

```python
model = Model(hidden_channels=32).to(device)
# Due to lazy initialization, we need to run one model step so the number
# of parameters can be inferred:
with torch.no_grad():
    model.encoder(train_data.x_dict, train_data.edge_index_dict)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
def train(train_data):
    model.train()
    optimizer.zero_grad()
    pred = model(train_data.x_dict, train_data.edge_index_dict,
                 train_data['user', 'movie'].edge_label_index)
    target = train_data['user', 'movie'].edge_label
    loss = F.binary_cross_entropy_with_logits(pred, target)
    loss.backward()
    optimizer.step()
    #pred=pred.round()
    #acc=accuracy_score(target,pred)
    return float(loss)
```

```python
@torch.no_grad()
def test(data):
    model.eval()
    pred = model(data.x_dict, data.edge_index_dict,
                 data['user', 'movie'].edge_label_index)
    pred = pred.clamp(min=0.0, max=1.0)
    target = data['user', 'movie'].edge_label.float()
    loss = F.binary_cross_entropy_with_logits(pred, target).sqrt()
    pred=pred.round()
    acc=accuracy_score(target,pred)
    return (float(loss),float(acc))
```

```python
train_l=[]
train_a=[]
val_l=[]
val_a=[]
test_l=[]
test_a=[]

for epoch in range(1, 301):
    train_loss = train(train_data)

    
    train_loss,train_acc = test(train_data)
    train_l.append(train_loss)
    train_a.append(train_acc)
    
    val_loss,val_acc = test(val_data)
    val_l.append(val_loss)
    val_a.append(val_acc)
    
    test_loss,test_acc = test(test_data)
    test_l.append(test_loss)
    test_a.append(test_acc)
    
    print(f'Epoch: {epoch:03d}, Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, | '
          f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, | Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f},')
```

```python
res=np.array(train_a)-np.array(val_a)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(8)
fig.set_figwidth(15)
fig.suptitle('SAGEConv Performance for Link Prediction on MovieLens')
ax1.plot(range(1,301),train_l,label='Train')
ax1.plot(range(1,301),val_l,label='Val')
ax1.plot(range(1,301),test_l,label='Test')
#ax1.title('Binary CrossEntropy Loss')
ax1.set_title('Binary CrossEntropy Loss')
ax1.set(xlabel='Iteration', ylabel='BCE Loss')
ax1.axvline(x = len(res[res<0.023]), color = 'r', label = 'Acc Diff <0.023')
ax1.legend()

ax2.plot(range(1,301),train_a,label='Train')
ax2.plot(range(1,301),val_a,label='Val')
ax2.plot(range(1,301),test_a,label='Test')
#ax1.title('Binary CrossEntropy Loss')
ax2.set_title('Accuracy')
ax2.set(xlabel='Iteration', ylabel='Accuracy')


ax2.axvline(x = len(res[res<0.023]), color = 'r', label = 'Acc Diff <0.023')
ax2.legend()
plt.savefig("SAGEConv_perf.png")
plt.show()
```

## Link Prediction using GAT

```python
class GATEncoder(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = GATConv((-1, -1), hidden_channels, add_self_loops=False)
        self.conv2 = GATConv((-1, -1), out_channels, add_self_loops=False)      
        
    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = self.conv2(x, edge_index) 
        return x
```

```python
class Model_GAT(torch.nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        self.encoder = GATEncoder(hidden_channels, hidden_channels)
        self.encoder = to_hetero(self.encoder, data.metadata(), aggr='sum')
        self.decoder = EdgeDecoder(hidden_channels)

    def forward(self, x_dict, edge_index_dict, edge_label_index):
        z_dict = self.encoder(x_dict, edge_index_dict)
        return self.decoder(z_dict, edge_label_index)
```
